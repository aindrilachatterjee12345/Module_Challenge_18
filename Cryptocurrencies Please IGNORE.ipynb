```python
# Remove CC that aren't trading

not_trading = crypto_df[crypto_df['IsTrading']==False].index
# print(not_trading)

crypto_df.drop(not_trading, inplace=True)

crypto_df.info()
```

```python
# Remove the IsTrading Column

crypto_df = crypto_df.drop(['IsTrading'],axis = 1)

crypto_df.head()
```

```python
# Remove CC's with at least 1 null value

crypto_df = crypto_df.dropna()

crypto_df.info()
```

```python
# Remove all CC without coins mined

not_mined = crypto_df[crypto_df['TotalCoinsMined']==0].index
# print(not_mined)

crypto_df.drop(not_mined, inplace=True)

crypto_df.info()
```

```python
# Store the names of all CC on a DataFramed named coins_name, and 
#use the crypto_df.index as the index for this new DataFrame

coins_name = pd.DataFrame(crypto_df['CoinName'])

coins_name.head()
```

```python
# Drop the CoinName column

crypto_df = crypto_df.drop(["CoinName"],axis=1)

crypto_df.head()
```

```python
# Create dummies variables for all of the text features (Algorithm and ProofType), and store the resulting data on a DataFrame named X.

X = pd.get_dummies(crypto_df, columns=["Algorithm", "ProofType"])

X.head()
```

```python
# Standardize data using StandardScaler

X = StandardScaler().fit_transform(X)

X
```

# Reducing Data Dimensions Using PCA¶

```python
#Use the PCA algorithm from sklearn (Links to an external site.) to reduce the dimensions of the X DataFrame down to three principal components.
Once you have reduced the data dimensions, create a DataFrame named “pcs_df” that includes the following columns: PC 1, PC 2, and PC 3. Use the crypto_df.index as the index for this new DataFrame.
```

```python
# Initialize PCA to 3 components

pca = PCA(n_components=3)

X_pca = pca.fit_transform(X)
```

```python
# create pcs_df

pca_df = pd.DataFrame(data=X_pca, columns = ["PC 1", "PC 2", "PC 3"], index = crypto_df.index)

pca_df.head()
```

# Clustering Cryptocurrencies Using K-means

```python
#Create an elbow curve to find the best value for K, and use the pcs_df DataFrame.
Once you define the best value for K, run the K-means algorithm to predict the K clusters for the cryptocurrencies’ data. Use the pcs_df to run the K-means algorithm.
Create a new DataFrame named “clustered_df,” that includes the following columns: Algorithm, ProofType, TotalCoinsMined, TotalCoinSupply, PC 1, PC 2, PC 3, CoinName, and Class.
```

```python
# Create empty inertia list and range of K values to test (copied from module)
inertia = []
k = list(range(1, 11))

# Loop through each K value, find inertia, and store in list
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pca_df)
    inertia.append(km.inertia_)
```
